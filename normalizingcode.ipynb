```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import math
from sklearn import preprocessing

# Set charts to view inline
%matplotlib inline
```

```python
df = pd.read_csv("fullsetmatrix1.csv")
x = df['1']
x
```

```python
# Create x, where x the 'scores' column's values as floats
y = df[['1']].values.astype(float)

# Create a minimum and maximum processor object
min_max_scaler = preprocessing.MinMaxScaler()

# Create an object to transform the data to fit minmax processor
y_scaled = min_max_scaler.fit_transform(y)

# Run the normalizer on the dataframe
df_normalized1 = pd.DataFrame(y_scaled)
df_normalized1
```

```python
# Create x, where x the 'scores' column's values as floats
z = df[['2']].values.astype(float)

# Create a minimum and maximum processor object
min_max_scaler = preprocessing.MinMaxScaler()

# Create an object to transform the data to fit minmax processor
z_scaled = min_max_scaler.fit_transform(z)

# Run the normalizer on the dataframe
df_normalized2 = pd.DataFrame(z_scaled)
df_normalized2
```

```python
# Create x, where x the 'scores' column's values as floats
a = df[['3']].values.astype(float)

# Create a minimum and maximum processor object
min_max_scaler = preprocessing.MinMaxScaler()

# Create an object to transform the data to fit minmax processor
a_scaled = min_max_scaler.fit_transform(a)

# Run the normalizer on the dataframe
df_normalized3 = pd.DataFrame(a_scaled)
df_normalized3
```

```python
# Create x, where x the 'scores' column's values as floats
b = df[['4']].values.astype(float)

# Create a minimum and maximum processor object
min_max_scaler = preprocessing.MinMaxScaler()

# Create an object to transform the data to fit minmax processor
b_scaled = min_max_scaler.fit_transform(b)

# Run the normalizer on the dataframe
df_normalized4 = pd.DataFrame(b_scaled)
df_normalized4
```

```python
# Create x, where x the 'scores' column's values as floats
c = df[['5']].values.astype(float)

# Create a minimum and maximum processor object
min_max_scaler = preprocessing.MinMaxScaler()

# Create an object to transform the data to fit minmax processor
c_scaled = min_max_scaler.fit_transform(c)

# Run the normalizer on the dataframe
df_normalized5 = pd.DataFrame(c_scaled)
df_normalized5
```

```python
result = pd.concat([df_normalized1, df_normalized2,df_normalized3,df_normalized4,df_normalized5], axis=1)
result
```

```python
result.columns = ['1', '2','3','4','5']
result
```

```python

result.to_csv(r'normalizedmatrix.csv')
```
